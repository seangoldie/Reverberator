```python
import reverberator
import soundfile as sf
import numpy as np
import IPython.display as ipd
```

```python
x, fs = sf.read("test_signal.wav")
```

```python
x.shape
```

```python
x_left, x_right = x.T
```

```python
x_left = np.append(x_left, np.zeros(int(0.5 * len(x_left))))
x_right = np.append(x_right, np.zeros(int(0.5 * len(x_right))))
```

```python
reverb = reverberator.SchroederReverberator(fs, num_combs=6, amount=0.75, delay_times=[20, 30, 40, 50, 60, 70, 25, 50])
```

```python
y = [reverb.process_buffer(x_left), reverb.process_buffer(x_right)]
```

```python
ipd.Audio(y, rate=fs)
```

```python
sf.write("output_signal.wav", np.array(y).T, samplerate=fs)
```
